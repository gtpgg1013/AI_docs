# Numpy로 계산해보장!

```
import numpy as np
timesteps = 10 # 시점, 단어의 숫자(한 문장), char의 수(한 단어)
# 주식 예측 모델 : timesteps = 7 => 일주일간의 주식 변동패턴을 기반으로 : 오늘 주식을 예측
input_dim = 4 # 단어가 4개인 벡터(corpus)
hidden_size = 8 # hidden vector의 크기 : cell의 크기
inputs = np.random.random((timesteps, input_dim))
inputs.shape
```

```
inputs
```

```
# 초기 은닉 상태를 0으로 초기화
hidden_state_t = np.zeros((hidden_size, ))
```

```
# 가중치
# wx : 입력에 대한 가중치
wx = np.random.random((hidden_size, input_dim))
# wh : 은닉층
wh = np.random.random((hidden_size, hidden_size))
# b : 바이어스
b = np.random.random((hidden_size,))
print(np.shape(wx))
```

```
# 메모리 셀 동작
total_hidden_states = []
for input_t in inputs:
  output_t = np.tanh(np.dot(wx,input_t) + np.dot(wh,hidden_state_t) + b)
  # wx*xt + wh+ht-1 + b
  total_hidden_states.append(list(output_t))
  # print(np.shape(total_hidden_states))
  hidden_state_t = output_t
print(total_hidden_states)
  
```

```
def word_gen(model, t, curr_word, num): # model, tokenizer, 입력단어, 개수
  ini_word = curr_word
  sent = ''
  for _ in range(num):
    encoded = t.texts_to_sequences([curr_word])[0] # 스페이스로 단어 구분해서 그걸 인덱스로 치환해서 리스트로 리턴
    print(encoded)
    encoded = pad_sequences([encoded], maxlen=max_len-1, padding='pre') # 그걸 사이즈 5(input size)로 변경해줌 : pad_sequences를 쓰면 맨 뒤에놈 5개만 가져와주는군! 개이득!
    print("변경!",encoded)
    res = model.predict_classes(encoded) # 다음 글자의 인덱스 번호!
    print(res)
    for word, index in t.word_index.items():
      if res==index:
        break
    curr_word = curr_word+' '+word
    sent = sent+' '+word
  sent = ini_word+sent
  return sent
    
    
print(word_gen(model, t, '멀티캠퍼스에', 6))
```

```
from string import punctuation
import re
def mypreprocessing(s): # 점 제거, 소문자 변환
  s = s.lower()
  tmp = ''
  for char in s:
    if char not in punctuation:
      tmp += char
  # 전처리된 문장 리턴
  print(''.join(ch for ch in s if ch not in punctuation))
  return tmp

text = [mypreprocessing(x) for x in headline]
```

```
# vacabulary 작성(단어집합)
t = Tokenizer()
t.fit_on_texts(text)
t.word_index
```

```
vocab_size = len(t.word_index) + 1 # 단어 집합의 크기
vocab_size
```

```
sequences = []
for line in text: # 1214개의 line
  encoded = t.texts_to_sequences([line])[0]
#   print(len(encoded))
  for i in range(1, len(encoded)):
    sequence = encoded[:i+1]
    sequences.append(sequence)
print(sequences[:20])
```

```
np.shape(sequences)
```

```
max_len = max(len(s) for s in sequences) # 최대길이 24
max_len
```

```
from keras.preprocessing.sequence import pad_sequences
sequences = pad_sequences(sequences,maxlen=max_len, padding='pre')
np.shape(sequences)
```

```
xdata = sequences[:,:-1]
ylabel = sequences[:,-1].reshape(-1,1)
y_ohe = to_categorical(ylabel)
```

```
model = Sequential()
model.add(Embedding(vocab_size, 10, input_length=max_len-1)) # 단어집합 크기(즉, 몇개?), 10차원으로, input length는 5 (6-1) = time step!
model.add(SimpleRNN(32))
model.add(Dense(vocab_size, activation='softmax'))
model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['acc'])
model.fit(xdata,y_ohe, epochs=150, verbose=2)
```
