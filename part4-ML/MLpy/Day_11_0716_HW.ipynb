```python
import numpy as np
import pandas as pd
import urllib
from bs4 import BeautifulSoup
import requests
from urllib import parse
```

```python
# 1번

base = "ko.wikisource.org/wiki/"
bn = "하늘과_바람과_별과_시/"
poemList = ['자화상','새로운 길','슬픈 족속','소년','병원','무서운 시간','눈 오는 지도',\
           '태초의 아침','또 태초의 아침','새벽이 올 때까지','십자가','눈감고 간다',\
           '돌아와 보는 밤','간판 없는 거리','바람이 불어','또 다른 고향','길',\
           '별 헤는 밤','서시']
resList = []
for poem in poemList:
    tempList = []
    site = base+bn+poem
    html = urllib.request.urlopen('https://'+parse.quote(site)).read().decode('utf-8')
    soup = BeautifulSoup(html, 'html.parser')
#     print(html)
#     print(site)
    try:
        tempList.append(soup.select('#header_section_text')[0].string)
    except:
        pass
    
    try:
        tempList.append(soup.select('#mw-content-text > div.mw-parser-output > div > p')[0].get_text())
    except:
        tempList.append(soup.select('#mw-content-text > div.mw-parser-output > p')[0].get_text())
    resList.append(tempList)    
    
                    
resList
```

```python
# 3번
url = 'https://stackoverflow.com/questions/tagged/python'
html = requests.get(url).text
soup = BeautifulSoup(html, 'html.parser')
```

```python
sof = 'https://stackoverflow.com'
# print(soup.select('div.question-summary'))
nurls = soup.select('a.question-hyperlink')
# print(len(nurls))
# print(nurls)
# nurls[0].attrs['href']
nurlList = []
for nurl in nurls:
    nurlList.append(nurl.attrs['href'])
```

```python
resList = []
for nurl in nurlList:
    tmpList = []
    site = sof + nurl
    print(site)
    html = requests.get(site).text
    soup = BeautifulSoup(html, 'html.parser')
    question = ''
    code = ''
    comment = ''
    answer = ''
    try:
        question += soup.select_one('#question > div.post-layout > div.postcell.post-layout--right >\
        div.post-text > p:nth-child(1)').string
        question += '\\n'
    except:
        pass
    
    try:
        question += soup.select_one('#question > div.post-layout > div.postcell.post-layout--right > div.post-text > blockquote > p').string
        question += '\\n'
    except:
        pass
    
    try:
        code += soup.select_one('#question > div.post-layout > div.postcell.post-layout--right > div.post-text > pre > code')
    except:
        pass
    
    try:
        comment += soup.select_one('div.comment-text.js-comment-text-and-form > div > span.comment-copy').string
    except:
        pass
    
    i = 1
    while True:
        try:
            answer += soup.select('div > div.answercell.post-layout--right > div.post-text > p:nth-child('+str(i)+')')
            answer += '\\n'
            i += 1
        except:
            break
            
    tmpList.append(question)
    tmpList.append(code)
    tmpList.append(comment)
    tmpList.append(answer)
    resList.append(tmpList)
```
