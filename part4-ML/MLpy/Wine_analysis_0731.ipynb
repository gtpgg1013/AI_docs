# Modeling

```python
x = tf.placeholder(dtype=tf.float32, shape=[None,8])
y = tf.placeholder(dtype=tf.int32, shape=[None,1])
#범주형 label one_hot encoding
y_ohe = tf.one_hot(y,7)
y_ohe = tf.reshape(y_ohe,[-1,7])
```

```python
w = tf.Variable(tf.random_normal([8,7]))
b = tf.Variable(tf.random_normal([7]))

logit = tf.matmul(x,w) + b
hf = tf.nn.softmax(logit)
# cost = tf.reduce_mean(-tf.reduce_sum(y_ohe*tf.log(hf), axis=1))
cost_ = tf.nn.softmax_cross_entropy_with_logits(logits=logit, labels=y_ohe)
cost = tf.reduce_mean(cost_)
train = tf.train.AdamOptimizer(0.01).minimize(cost)

prediction = tf.argmax(hf,1)
accuracy = tf.reduce_mean(tf.cast(tf.equal(prediction,tf.argmax(y_ohe,1)),\
                                  dtype=tf.float32))
```

```python
with tf.Session() as sess:
    sess.run(tf.global_variables_initializer())
    for step in range(4001):
        sess.run(train, feed_dict={x:x_train, y:y_train})
        if step%400==0:
            cv, acc = sess.run([cost, accuracy], feed_dict={x:x_train, y:y_train})
            print("cv: {}, acc: {}".format(cv,acc))
    pred, real, acc_pred = sess.run([prediction, y, accuracy], \
                               feed_dict={x:x_test, y:y_test})
    print("="*50)
    for p, y in zip(pred, real):
        print("[{}] 예측: {}, 실제: {}".format(p==y, p, y))
    print("accuracy: {}".format(acc_pred))
```

```python
w = tf.Variable(tf.random_normal([num_features,7]))
b = tf.Variable(tf.random_normal([7]))

logit = tf.matmul(x,w) + b
hf = tf.nn.softmax(logit)
# cost = tf.reduce_mean(-tf.reduce_sum(y_ohe*tf.log(hf), axis=1))
cost_ = tf.nn.softmax_cross_entropy_with_logits(logits=logit, labels=y_ohe)
cost = tf.reduce_mean(cost_)
train = tf.train.AdamOptimizer(0.01).minimize(cost)

prediction = tf.argmax(hf,1)
accuracy = tf.reduce_mean(tf.cast(tf.equal(prediction,tf.argmax(y_ohe,1)),\
                                  dtype=tf.float32))
```

```python
with tf.Session() as sess:
    sess.run(tf.global_variables_initializer())
    for step in range(4001):
        sess.run(train, feed_dict={x:x_train, y:y_train})
        if step%400==0:
            cv, acc = sess.run([cost, accuracy], feed_dict={x:x_train, y:y_train})
            print("cv: {}, acc: {}".format(cv,acc))
    pred, real, acc_pred = sess.run([prediction, y, accuracy], \
                               feed_dict={x:x_test, y:y_test})
    print("="*50)
    for p, y in zip(pred, real):
        print("[{}] 예측: {}, 실제: {}".format(p==y, p, y))
    print("accuracy: {}".format(acc_pred))
```
