## class -> label Encoding

```python
from sklearn.preprocessing import LabelEncoder
enc_sum = LabelEncoder()
sum_label = enc_sum.fit_transform(train1.Summary)
```

```python
sum_label
```

```python
train1.Summary = sum_label
```

```python
enc_pre = LabelEncoder()
precip_label = enc_pre.fit_transform(train1['Precip Type'])
```

```python
precip_label
```

```python
train1['Precip Type'] = precip_label
```
