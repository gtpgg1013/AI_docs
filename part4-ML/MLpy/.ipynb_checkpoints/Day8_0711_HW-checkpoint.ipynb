```python
from string import ascii_lowercase
alpha_list = list(ascii_lowercase)
alpha_list
```

```python
pd.DataFrame(data=[[0]*26],columns=alpha_list)
```

```python
def count_letter(name):
#     name = name.values
#     print(name)
    res = pd.DataFrame(data=[[0]*26],columns=alpha_list) # 기본 시리즈 하나 만드러 주고
    for char in name:
        char = char.lower()
        res[char] += 1
#     print(res)
    return res.values # values로 하면 값들이 얻어지나?

# count_letter("aDDFSD")
```

# 메커니즘을 기억하면 좋다
## 다른 자료구조가 필요할 때는
## 열 별로 정리해서(list, dictionary)
## list에 넣어준 후에,
## list 전체를 pd.DataFrame으로 만들면 간단하다!

```python
tot_year = []
for name in map(str.lower, sum_name.values):
    char_dict = {}
    for char in name:
        if char in char_dict:
            char_dict[char] += 1
        else:
            char_dict[char] = 1
    tot_year.append(char_dict)

totyear = pd.DataFrame(tot_year, index=range(1880,2011))
totyear.head()
```

```python
totyear.sum().sort_values()
```

```python
totyear.head()
```

# 영화 데이터 처리

```python
unames = ['user_id','gender','age','occupation','zip']
rnames = ['user_id','movie_id', 'rating', 'timestamp']
mnames = ['movie_id', 'title', 'genres']
```

```python
users=pd.read_csv("data_0705/movielens/users.dat",
                  names=unames,header=None,sep="::")
ratings=pd.read_csv("data_0705/movielens/ratings.dat",
                    names=rnames,header=None,sep="::")
movies=pd.read_csv("data_0705/movielens/movies.dat",
                   names=mnames,header=None,sep="::")
```

```python
ratings.head()
```

```python
movies.head()
```

```python
users.head()
```

```python
import re
# map함수 써줄때는 하나의 놈에 해줄 수 있는걸 해주면댐!
def regex(string):
    pat = re.compile('\((\d+)\)')
    res = pat.search(string).group(1)
    return res
# pat.search('slfnsdl (202)').group(1)

movies['year'] = movies['title'].map(regex)
movies
```

```python
mg = movies.genres.str.split('|') # 이렇게 쉽게 처리하면 된다!
mg.head()
```

```python
pd.Series(mg[0]).head()
# 아하 이런식으로도 만들 수 있네
# => mg라는 Serise를 일종의 array index처럼 다룰 수 있음!
```

```python
# 요부분이 중요하네
# 각 COL마다 쫙 나눠주고
mg = mg.apply(lambda x:pd.Series(x))
mg.head()
```

```python
mg.stack() # 그걸 stack()로 쫙 풀어주고
```

```python
mg.stack().reset_index(level=1,drop=True)
# 촥 풀어준걸 인덱스를 다시 리셋해주고 (멀티인덱스 => 레벨1로 1줄로!)
```

```python
mg.stack().reset_index(level=1,drop=True).unique()
# 이러면 영화제목 뭐있는지 완성!
```
