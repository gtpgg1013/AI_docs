```python
years = range(1880, 2011)
pieces = []
for year in years:
    path = "babynames/yob{}.txt".format(year)
    frame = pd.read_csv(path, names=['name','sex','births'])
    frame['year'] = year
    frame = frame.sort_values(by='births', ascending=False) # 여기서 미리 sort를 해주자!
    pieces.append(frame)
names = pd.concat(pieces, ignore_index=True)
names.shape
```

```python
names.info()
```

```python
names.head()
```

```python
def add_prop(group): # dataframe을 받아서 col을 하나 더 만들어서 리턴해주는 함수
    group['prop'] = group.births / group.births.sum()
    return group

names_g = names.groupby('year').apply(add_prop)

# names_g는 이름을 year로 그룹화해서, 확률 row를 만든!
```

```python
names_g[names_g['year']==1880].prop.sum()
# 그니까 결론 => dataframe을 return하는 함수를 만들고 그거를 apply해주면 결과도 df가 나온다?
```

```python
# 과거에는 인기가 없었으나, 요즘에 떡상한 이름?
# 어떻게 찾는게 좋을까?
names_gp = names_g.groupby('year').apply(lambda x: x.sort_values(by='prop',\
                                                                ascending=False))
names_gp

######################################################################################
######################## 위의 names_gp를 만들 필요가 없었음 ##############################
######################################################################################

# rank 만들어주는 간단한 함수!
def makerank(data):
    data['rank'] = range(1,len(data)+1)
    return data

# makerank(names_gp[names_gp['year']==1880])
# 전체가 1인가 확인하려고 쳐봄

def add_prop(group): # dataframe을 받아서 col을 하나 더 만들어서 리턴해주는 함수
    group['prop'] = group.births / group.births.sum()
    return group

# name_g는 일단 prop라는 row를 하나 더 만든 DF!
names_g = names.groupby('year').apply(add_prop)

# 그놈을 다시 year로 그룹화해서 rank 붙여줌 ! => 근데 왜 정렬되있는진 모르겠는데? 정렬시켜줘야함
result = names_g.groupby('year').apply(makerank)
result.head(10)
```

```python
# 좋아! 이제 각 그룹별 랭크를 구했으니 랭크변화가 큰 놈을 찾아보자!
# 이름으로 그룹화 해서 => 전기 랭크값의 합 - 후기 랭크값의 합 큰 순으로 정렬!
frontRankVal = result[result['year']<1950].groupby('name')['rank'].sum()
frontRankVal.head()
frontRankVal.shape
```

```python
import json
with open("data_0705/bitly_usagov/example.txt",'r') as fr:
    print(fr.readline())

records = [json.loads(line) for line in open("data_0705/bitly_usagov/example.txt",\
                                            'r', encoding='utf-8')]
time_zones = [rec['tz'] for rec in records if 'tz' in rec]

def get_counts(seq):
    counts = {}
    for s in seq:
        if s in counts:
            counts[s] += 1
        else:
            counts[s] = 1
    return counts
```
