# 07/12 주요함수 복습

```python
years = range(1880, 2011)
pieces = []
for year in years:
    path = "babynames/yob{}.txt".format(year)
    frame = pd.read_csv(path, names=['name','sex','births'])
    frame['year'] = year
    frame = frame.sort_values(by='births', ascending=False) # 여기서 미리 sort를 해주자!
    pieces.append(frame)
names = pd.concat(pieces, ignore_index=True)
names.shape
```

```python
names.head()
```

```python
def makenameprop_rank(data):
    data['prop'] = data['births']/data['births'].sum()
    data['rank'] = range(1,len(data)+1)
    return data
```

```python
name_g = names.groupby('year').apply(makenameprop_rank)
name_g.head()
```

```python
# 이제 rank를 찾았으니 뭐가 떡상 떡락했는지 보자
frontyear = name_g[name_g['year']<1945]
```

```python
backyear = name_g[name_g['year']>=1945]
backyear.head()
```

```python
frontscore = frontyear.groupby('name')['rank'].sum() 
frontscore.head()
# frontyear 사람들의 rank점수 : 어찌되욌든 표준화만 하면 됨 ㅋㅋ
```

```python
backscore = backyear.groupby('name')['rank'].sum()
backscore.head()
# backyear 사람들의 rank점수 : 어찌되욌든 표준화만 하면 됨 ㅋㅋ
```

```python
# 자 이제 두개를 어떻게 합칠 것이냐?
maintained = pd.merge(frontscore, backscore, on='name')
maintained.head()
# 중요한 점 : 이렇게 하면 아예 사라진 이름은 없음!
```

```python
maintained['finalscore'] = maintained['rank_x'] - maintained['rank_y']
```

```python
maintained = maintained.sort_values(by='finalscore')
from sklearn.preprocessing import MinMaxScaler, StandardScaler
scaler =  StandardScaler()
scaled_data = scaler.fit(maintained['finalscore'].values.reshape(-1,1))
scaled_data = scaler.transform(maintained['finalscore'].values.reshape(-1,1))
scaled_data = scaled_data.reshape(-1,)
scaled_df = pd.Series(scaled_data, index=maintained.index)
scaled_df.plot()
# 이거슨 정규화된 자료로 그린(별 의미없는) 그래프
# maintained_corrupted = maintained.head(10)
# maintained_higher = maintained.tail(10)
```

```python
scaled_df.head(10).plot(kind='bar') # 요놈이 떡락스
import matplotlib.pyplot as plt
plt.ylim(-4.8,-4)
```
