```python
# 전처리 : 소문자, stopword(불용어사전), punctuations(컴마, 대괄호, 특수문자), 숫자 등 
def word_preprocess(word):
    word = word.lower()
    # 불용어 사전 정의
    # "의미가 없는" 것으로 보고 삭제할 것임
    stop_words = ['a', 'an', 'the', 'in', 'with', 'to', 'for', 'from', 'of', 'at', 'on', 'until', 'by', 'and', 'but', 'is', 'are', 'was', 'were', 'it', 'that', 'this', 'my', 'his', 'her', 'our', 'as', 'not'] # make your own list
    symbols = [',', '.', ':', '-', '+', '/', '*', '&', '%', '[', ']', '(', ')'] # make your own list
    for stop_word in stop_words:
        if word != stop_word:
            word = word
        else:
            word = '' # 날려버리기
    for symbol in symbols:
        word = word.replace(symbol,'') # 날려버리기 # replace returnn 넘겨줘야 함
        
    return word
```

```python
token_inx = {}
file_opened = open('python_wikipedia.txt', mode='r')
for line in file_opened.readlines():
    line.strip() # 왼, 오 공백 제거됨
#     print(line)
    for word in line.split():
        word = word_preprocess(word)
        if word not in token_inx: # word가 토큰 딕셔너리에 없다면
            if word != '': # 그리고 그 word가 빈칸이 아니라면
                token_inx[word] = 1 # 초깃값 설정
        else:
            token_inx[word] += 1
        # print(word)
# {python : 200, programming : 50... ] : 딕셔너리 구성하고픔
print(token_inx) # 단어개수 세기 완료!
```

```python
# binomial : 이항분포로부터 무작위 표본 추출
# 이산형 / 연속형 확률 분포
# 이산형(이항, 포아송)
# 연속형(정규, t분포, 균등, f, 카이제곱)
# 성공(1), 실패(0)
# 성공확률을 p인 베르누이 시행을 n번 수행, 성공하는 횟수 X
# => 확률변수 X는 모수 n과 p인 이항분포를 따른다
# f(x) = nCx * p^x * (1-p)^(n-x), x=0,1,2,...,n
```

```python
import os
import urllib.request

# raw_data_urls.txt : 다운로드 url들이 작성된 파일 : 이용해서 다운받는 법!
with open("data_0705/raw_data_urls.txt","r") as data_urls:
#     data_urls
#     print(data_urls)
    for line, url in enumerate(data_urls): # 순서로 만들어주는 함수 enum
        if line==2:
            break
        fn = url.split("/")[-1].strip() # 원래 다운로드 링크
        fp = os.path.join('','./data_0705',fn) # 내려받을 곳
#         print(fp)
        urllib.request.urlretrieve(url,fp) # 다운로드 할 때 사용하는 함수
        print("file "+str(line)+" download complete")
```
