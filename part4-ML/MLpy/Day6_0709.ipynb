```python
# contour / contourf : 등고선, 등고선 color
def myFunc(x,y):
    return (1-x/2 + x**4 + y**2)*np.exp(-x**2-y**2)

x = np.linspace(-3,3,256)
y = np.linspace(-3,3,256)
xx, yy = np.meshgrid(x,y) 
# 등고선을 그리고자하는 대상 데이터를 나타내는 특별한 행렬
zz = myFunc(xx,yy)
zz
```

```python
# x, y의 의미:
# z = height value
plt.contourf(xx,yy,zz,alpha=.75, cmap='jet')
# plt.contour(xx,yy,zz,alpha=.75)
```
