```python
import math
import numpy as np
```

```python
def sigmoid(z):
    return 1/(1+math.e**-z)
```

```python
def costFunc(W,X,y):
    m = y.size
    h = sigmoid(np.dot(W,X))
    cost = -(1/m) * sum(y*np.log(h) + (1-y)*np.log(1-h))
    grad = (1/m) * np.dot(X, h-y)
    return cost, grad
```
