```python
import numpy as np
import pandas as pd

mapping = {"Mr" : 0, "Miss" : 1, "Mrs" : 2}

data = pd.DataFrame({
    'title' : ['Mr','Miss','Mrs']
})

data['OHE'] = data['title'].map(mapping) # 딕셔너리를 map 함수의 인자로 던져줄 수도 있음
data
```

```python
train = pd.read_csv("titanic_data/train.csv")
test = pd.read_csv("titanic_data/test.csv")
train_test_data = [train,test]
for data in train_test_data:
    # 정규표현식 사용 => 호칭 추출 => 호칭별 건수 출력
    data['title'] = data['Name'].str.findall('.*, ([^.\s]*)')
#     print(data['title'])s
#     print(data.title.map(lambda x:x[0]))
    data['title'] = data['title'].map(lambda x :x[0])
#     print(data['title'])
    print(data.title.value_counts())
    
for data in train_test_data:
    data['OHE'] = np.where(data['title'].values=='Mr',0,
                          np.where(data['title'].values=='Miss',1,
                                  np.where(data['title'].values=='Mrs',2,3)))
# 지난 시간
```

```python
X_train = train[fn]
print(X_train.head())
Y_label = train['Survived']
Y_label.head()
```

```python
X_test = test[fn]
```

```python
from sklearn.tree import DecisionTreeClassifier # 분류니까
model = DecisionTreeClassifier(max_depth=3)
model
# depth 3이면 : 기준노드(1) + 그밑(2) + 그밑(3) : 연산량이 커서 컬럼 3개만 보겠다
```

```python
model.fit(X_train, Y_label) # 학습은 그냥 데이터 넣어주면 됨
```

```python
from sklearn.model_selection import KFold, cross_val_score
# k-fold 교차검증
# 가지고 있는 자체 데이터로 검증하다
kfold = KFold(n_splits=10, shuffle=True, random_state=2019) # 난수는 걍 숫자주면 됨
kfold # fold갯수 default 3
```

```python
score = cross_val_score(model, X_train, Y_label, cv=kfold, scoring='accuracy')
# 만든 모델, X데이터, y라벨, cv=kfold(만든 객체), scoring=정확도
```

```python
print(score)
```

```python
round(np.mean(score)*100,2)
```

```python
# 이제 실제로 예측해보자
prediction = model.predict(X_test)
prediction
```
